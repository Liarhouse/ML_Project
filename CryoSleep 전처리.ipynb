# CryoSleep 결측값 처리를 위한 XGBoost 분류

```python
df_new.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_new.drop(['CryoSleep', 'combi', 'Cabin', 'Transported'],axis=1),df_new.CryoSleep,
                                                    random_state=109)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train)
```

```python
import xgboost as xgb

trdm = xgb.DMatrix(data=X_train, label=y_train)
tdm = xgb.DMatrix(data=X_test, label=y_test)
vdm = xgb.DMatrix(data=X_val, label=y_val)
```

```python
params = {'max_depth':3,
         'eta':0.05,
         'objective':'binary:logistic',
         'eval_metric':'logloss'
         }
num_rounds = 400
```

```python
eval_list = [(trdm, 'train'), (vdm, 'eval')]

xgb_model = xgb.train(params = params, dtrain=trdm, num_boost_round=num_rounds,
                     early_stopping_rounds=50, evals=eval_list)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, roc_auc_score, f1_score

def  get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    roc_auc = roc_auc_score(y_test, pred_proba)
    
    print('오차 행렬')
    print(confusion)
 
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, \
    재현율: {2:.4f}, F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

## 훈련셋 결과

```python
train_pred_probs = xgb_model.predict(trdm)
print('predict() 수행 결과값을 10개만 표시, 예측 확률 값으로 표시됨')
print(np.round(train_pred_probs[:10],3))

train_preds = [1 if x > 0.5 else 0 for x in train_pred_probs]
print('예측값 10개만 표시:', train_preds[:10])
```

```python
get_clf_eval(y_train, train_preds, train_pred_probs)
```

## 테스트셋 결과

```python
pred_probs = xgb_model.predict(tdm)
print('predict() 수행 결과값을 10개만 표시, 예측 확률 값으로 표시됨')
print(np.round(pred_probs[:10],3))

preds = [1 if x > 0.5 else 0 for x in pred_probs]
print('예측값 10개만 표시:', preds[:10])
```

```python
get_clf_eval(y_test, preds, pred_probs)
```

## 검증셋 결과

```python
v_pred_probs = xgb_model.predict(vdm)
print('predict() 수행 결과값을 10개만 표시, 예측 확률 값으로 표시됨')
print(np.round(v_pred_probs[:10],3))

v_preds = [1 if x > 0.5 else 0 for x in v_pred_probs]
print('예측값 10개만 표시:', v_preds[:10])
```

```python
get_clf_eval(y_val, v_preds, v_pred_probs)
```
